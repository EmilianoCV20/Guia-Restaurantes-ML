```python
#TOP RESULTADOS CON FILTRO (REPETICION)
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, max

# Inicializar SparkSession
spark = SparkSession.builder \
    .appName("TopRestaurantePorNombre") \
    .getOrCreate()

# Ruta del archivo CSV
file_path = './Datasets/InspeccionRestaurantesNYC.csv'

# Leer los datos
data = spark.read.csv(file_path, header=True, inferSchema=True)

# Convertir la columna INSPECTION DATE a formato de fecha
data = data.withColumn("INSPECTION DATE", col("INSPECTION DATE"))

# Filtrar por las fechas y puntajes relevantes
filtered_data = data.filter(
    (col("INSPECTION DATE") >= "01/01/2014") &
    (col("INSPECTION DATE") <= "12/31/2024") &
    (col("SCORE") >= 75) & (col("SCORE") <= 170)
).select("DBA", "BORO", "INSPECTION DATE", "SCORE", "Latitude", "Longitude")

# Agrupar por nombre (DBA) y seleccionar la máxima puntuación (SCORE)
best_per_restaurant = filtered_data.groupBy("DBA").agg(
    max("SCORE").alias("MAX_SCORE")
).alias("best")

# Alias para el DataFrame filtrado
filtered_data_alias = filtered_data.alias("filtered")

# Unir los DataFrames utilizando columnas calificadas explícitamente
best_restaurants = best_per_restaurant.join(
    filtered_data_alias,
    (col("best.DBA") == col("filtered.DBA")) & 
    (col("best.MAX_SCORE") == col("filtered.SCORE")),
    "inner"
).select(
    col("filtered.DBA").alias("DBA"),
    col("filtered.BORO").alias("BORO"),
    col("filtered.INSPECTION DATE").alias("INSPECTION_DATE"),
    col("filtered.SCORE").alias("SCORE"),
    col("filtered.Latitude").alias("Latitude"),
    col("filtered.Longitude").alias("Longitude")
)

# Mostrar resultados
best_restaurants.show(truncate=False)

# Finalizar la sesión Spark
spark.stop()
```

```python
#TOP RESULTADOS YA CON FILTRO (SIN REPETICION)
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, dense_rank
from pyspark.sql.window import Window

# Inicializar SparkSession
spark = SparkSession.builder \
    .appName("TopRestaurantesUnicos") \
    .getOrCreate()

# Ruta del archivo CSV
file_path = './Datasets/InspeccionRestaurantesNYC.csv'

# Leer los datos
data = spark.read.csv(file_path, header=True, inferSchema=True)

# Convertir la columna INSPECTION DATE a formato de fecha
data = data.withColumn("INSPECTION DATE", col("INSPECTION DATE"))

# Filtrar por las fechas y puntajes relevantes
filtered_data = data.filter(
    (col("INSPECTION DATE") >= "01/01/2014") &
    (col("INSPECTION DATE") <= "12/31/2024") &
    (col("SCORE") >= 75) & (col("SCORE") <= 170)
).select("DBA", "BORO", "INSPECTION DATE", "SCORE", "Latitude", "Longitude")

# Eliminar duplicados basados en DBA, BORO, Latitude y Longitude
unique_data = filtered_data.dropDuplicates(["DBA", "BORO", "Latitude", "Longitude"])

# Crear una ventana global para rankear por puntuación y seleccionar el top 50
global_window = Window.orderBy(col("SCORE").desc())
top_50_restaurants = unique_data.withColumn("global_rank", dense_rank().over(global_window)) \
    .filter(col("global_rank") <= 50) \
    .select("DBA", "BORO", "INSPECTION DATE", "SCORE", "Latitude", "Longitude")

# Mostrar resultados
top_50_restaurants.show(truncate=False)

# Finalizar la sesión Spark
spark.stop()
```

```python
#TOP RESTAURANTES POR CADA ZONA DE NY
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window

# Inicializar SparkSession
spark = SparkSession.builder \
    .appName("TopRestaurantesPorBoro") \
    .getOrCreate()

# Ruta del archivo CSV
file_path = './Datasets/InspeccionRestaurantesNYC.csv'

# Leer los datos
data = spark.read.csv(file_path, header=True, inferSchema=True)

# Convertir la columna INSPECTION DATE a formato de fecha
data = data.withColumn("INSPECTION DATE", col("INSPECTION DATE"))

# Filtrar por fechas y puntajes relevantes
filtered_data = data.filter(
    (col("INSPECTION DATE") >= "01/01/2014") &
    (col("INSPECTION DATE") <= "12/31/2024") &
    (col("SCORE") >= 75) & (col("SCORE") <= 170)
).select("DBA", "BORO", "INSPECTION DATE", "SCORE", "Latitude", "Longitude")

# Eliminar duplicados por DBA, BORO, Latitude, y Longitude, conservando la puntuación más alta
window_spec = Window.partitionBy("DBA", "BORO", "Latitude", "Longitude").orderBy(col("SCORE").desc())
unique_restaurants = filtered_data.withColumn("rank", row_number().over(window_spec)) \
    .filter(col("rank") == 1) \
    .drop("rank")

# Crear una ventana para el ranking por cada BORO
boro_window = Window.partitionBy("BORO").orderBy(col("SCORE").desc())

# Agregar un ranking dentro de cada BORO
ranked_data = unique_restaurants.withColumn("boro_rank", row_number().over(boro_window))

# Filtrar el Top 50 de cada BORO
top_per_boro = ranked_data.filter(col("boro_rank") <= 50).drop("boro_rank")

# Mostrar el resultado por BORO
for boro in top_per_boro.select("BORO").distinct().collect():
    boro_name = boro["BORO"]
    print(f"Top restaurantes en {boro_name}:\n")
    top_per_boro.filter(col("BORO") == boro_name).show(truncate=False)

# Finalizar la sesión Spark
spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window
import pandas as pd
import folium

# Inicializar SparkSession
spark = SparkSession.builder \
    .appName("TopRestaurantesPorBoro") \
    .getOrCreate()

# Ruta del archivo CSV
file_path = './Datasets/InspeccionRestaurantesNYC.csv'

# Leer los datos
data = spark.read.csv(file_path, header=True, inferSchema=True)

# Convertir la columna INSPECTION DATE a formato de fecha
data = data.withColumn("INSPECTION DATE", col("INSPECTION DATE"))

# Filtrar por fechas y puntajes relevantes
filtered_data = data.filter(
    (col("INSPECTION DATE") >= "01/01/2014") &
    (col("INSPECTION DATE") <= "12/31/2024") &
    (col("SCORE") >= 75) & (col("SCORE") <= 170)
).select("DBA", "BORO", "INSPECTION DATE", "SCORE", "Latitude", "Longitude")

# Eliminar duplicados por DBA, BORO, Latitude, y Longitude, conservando la puntuación más alta
window_spec = Window.partitionBy("DBA", "BORO", "Latitude", "Longitude").orderBy(col("SCORE").desc())
unique_restaurants = filtered_data.withColumn("rank", row_number().over(window_spec)) \
    .filter(col("rank") == 1) \
    .drop("rank")

# Crear una ventana para el ranking por cada BORO
boro_window = Window.partitionBy("BORO").orderBy(col("SCORE").desc())

# Agregar un ranking dentro de cada BORO
ranked_data = unique_restaurants.withColumn("boro_rank", row_number().over(boro_window))

# Filtrar el Top 50 de cada BORO
top_per_boro = ranked_data.filter(col("boro_rank") <= 50).drop("boro_rank")

# Convertir los datos Spark a Pandas para trabajar con Folium
top_per_boro_pd = top_per_boro.toPandas()

# Generar mapas por BORO
for boro in top_per_boro_pd["BORO"].unique():
    # Filtrar datos para el BORO actual
    boro_data = top_per_boro_pd[top_per_boro_pd["BORO"] == boro]
    boro_data = boro_data.dropna(subset=["Latitude", "Longitude"])  # Eliminar NaN en ubicaciones
    
    if boro_data.empty:
        print(f"No hay datos válidos para {boro}.")
        continue
    
    # Crear un mapa centrado en el promedio de las ubicaciones
    mapa = folium.Map(
        location=[boro_data["Latitude"].mean(), boro_data["Longitude"].mean()],
        zoom_start=12
    )
    
    # Agregar marcadores al mapa
    for _, row in boro_data.iterrows():
        folium.Marker(
            location=[row["Latitude"], row["Longitude"]],
            popup=f"{row['DBA']} (Score: {row['SCORE']})",
            tooltip=row["DBA"],
            icon=folium.Icon(color="blue", icon="cutlery", prefix="fa")
        ).add_to(mapa)
    
    # Guardar el mapa como archivo HTML
    mapa.save(f"mapa_{boro.lower()}.html")
    print(f"Mapa generado para {boro}: mapa_{boro.lower()}.html")

# Finalizar la sesión Spark
spark.stop()
```
