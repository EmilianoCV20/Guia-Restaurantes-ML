```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#df = pd.read_csv("./Datasets/InspeccionRestaurantesNYC_15Oct2024.csv")
#df = df.drop('Location Point1', axis=1)
df = pd.read_csv("./Datasets/InspeccionRestaurantesNYC_15Oct2024.csv")
df = df.drop('Location Point1', axis=1)
df['DBA'] = df['DBA'].fillna('FOOD CART')
df['BUILDING'] = df['BUILDING'].fillna(0)
df['ZIPCODE'] = df['ZIPCODE'].fillna(10001)
df['PHONE'] = df['PHONE'].fillna(9148152703)
df['CUISINE DESCRIPTION'] = df['CUISINE DESCRIPTION'].fillna('Fast Food')
df['ACTION'] = df['ACTION'].fillna('No violations were recorded at the time of this inspection.')
df['VIOLATION CODE'] = df['VIOLATION CODE'].fillna('0N')
df['VIOLATION DESCRIPTION'] = df['VIOLATION DESCRIPTION'].fillna('None.')
df['SCORE'] = df['SCORE'].fillna(0)
df['GRADE'] = df['GRADE'].fillna('N')
df['GRADE DATE'] = df['GRADE DATE'].fillna('08/15/2024')
df['INSPECTION TYPE'] = df['INSPECTION TYPE'].fillna('Cycle Inspection / Initial Inspection')
df['Latitude'] = df['Latitude'].fillna(0)
df['Longitude'] = df['Longitude'].fillna(0)
df['Community Board'] = df['Community Board'].fillna(0)
df['Council District'] = df['Council District'].fillna(0)
df['Census Tract'] = df['Census Tract'].fillna(0)
df['BIN'] = df['BIN'].fillna(0)
df['BBL'] = df['BBL'].fillna(0)
df['NTA'] = df['NTA'].fillna('NA00')

# 1. Seleccionar las columnas para el modelo (columnas numéricas)
df_selected = df[['SCORE', 'Latitude', 'Longitude', 'ZIPCODE', 'GRADE']]

# 2. Codificar las columnas categóricas
df_selected = pd.get_dummies(df_selected, columns=['GRADE'])

# 3. Convertir a arrays de NumPy (Separar las características y las etiquetas)
labels = np.array(df_selected['SCORE'])

# Eliminar las etiquetas de las características
features = df_selected.drop('SCORE', axis=1)

# Guardar los nombres de las características para referencia futura
feature_list = list(features.columns)

# Convertir las características a un array de NumPy
features = np.array(features)

# 4. Dividir los datos en conjuntos de entrenamiento y prueba (75% entrenamiento, 25% prueba)
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=0.25, random_state=42
)

# Imprimir las dimensiones de los conjuntos de entrenamiento y prueba
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# 5. Normalizar las características
scaler = StandardScaler()

# Ajustar el escalador a los datos de entrenamiento y luego transformar
train_features = scaler.fit_transform(train_features)

# Usar el mismo escalador para transformar el conjunto de prueba
test_features = scaler.transform(test_features)

# Ahora los conjuntos de datos están listos para ser usados en un modelo de machine learning.
from sklearn.linear_model import LinearRegression

# Crear un modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo con los datos de entrenamiento
model.fit(train_features, train_labels)

# Hacer predicciones con los datos de prueba
predictions = model.predict(test_features)

# Evaluar el modelo (por ejemplo, usando R^2 o error cuadrático medio)
from sklearn.metrics import mean_squared_error, r2_score

# Calcular el error cuadrático medio y el R^2
mse = mean_squared_error(test_labels, predictions)
r2 = r2_score(test_labels, predictions)

# Imprimir las métricas del modelo
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
# Calculamos el promedio de la etiqueta (SCORE) en el conjunto de entrenamiento
baseline_preds = np.mean(train_labels)

# Utilizamos este valor como la predicción para todas las instancias en el conjunto de prueba
baseline_preds = np.full(test_labels.shape, baseline_preds)

# Calculamos el error absoluto entre las predicciones base y las etiquetas reales
baseline_errors = abs(baseline_preds - test_labels)

# Mostramos el error promedio
print('Average baseline error: ', round(np.mean(baseline_errors), 2))
```

```python
# Importar el modelo RandomForestRegressor de sklearn
from sklearn.ensemble import RandomForestRegressor

# Instanciar el modelo con 1000 árboles de decisión
rf = RandomForestRegressor(n_estimators=1000, random_state=42)

# Entrenar el modelo con los datos de entrenamiento
rf.fit(train_features, train_labels)

# Realizar predicciones en el conjunto de prueba
predictions = rf.predict(test_features)

# Calcular los errores absolutos
errors = abs(predictions - test_labels)

# Imprimir el error absoluto medio (MAE)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
# Crear un nuevo modelo de bosque aleatorio con solo las dos características más importantes
rf_most_important = RandomForestRegressor(n_estimators=1000, random_state=42)

# Asegúrate de que las dos características más importantes estén en el 'feature_importances'
most_important_features = [feature_importances[0][0], feature_importances[1][0]]

# Obtener los índices de las dos características más importantes en 'feature_list'
important_indices = [feature_list.index(most_important_features[0]), feature_list.index(most_important_features[1])]

# Extraer esas características para los conjuntos de entrenamiento y prueba
train_important = train_features[:, important_indices]
test_important = test_features[:, important_indices]

# Entrenar el bosque aleatorio con las dos características más importantes
rf_most_important.fit(train_important, train_labels)

# Hacer predicciones con el modelo entrenado
predictions = rf_most_important.predict(test_important)

# Calcular los errores absolutos
errors = abs(predictions - test_labels)

# Imprimir el error absoluto medio (MAE)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')

# Filtrar los valores de test_labels que no son cero para calcular el MAPE
non_zero_test_labels = test_labels[test_labels != 0]
non_zero_errors = errors[test_labels != 0]

# Calcular el MAPE solo con los valores no nulos
if len(non_zero_test_labels) > 0:
    mape = np.mean(100 * (non_zero_errors / non_zero_test_labels))
    accuracy = 100 - mape
    print('Accuracy:', round(accuracy, 2), '%.')
else:
    print('No non-zero test labels to calculate accuracy.')
```
